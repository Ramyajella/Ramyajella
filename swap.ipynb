```python
import cv2
import numpy as np
```

```python
image1 = cv2.imread('nature.jpg')
```

```python
cv2.imshow("nature", image1)
cv2.waitKey() == 12
cv2.destroyAllWindows()
```

```python
image2 = cv2.imread('moana.jpg')
```

```python
cv2.imshow("moana", image2)
cv2.waitKey() == 12
cv2.destroyAllWindows()
```

```python
image1 = image1[:720, :1280]
```

```python
image1.shape
```

```python
cphoto1 = image2[120:720, :]
cphoto2 = image1[200:350, :]
cphoto3 = image1[150:220, :]
cphoto4 = np.concatenate((cphoto3,cphoto1,cphoto2))
```

```python
cv2.imshow("final_img",cphoto4 )
cv2.waitKey() 
cv2.destroyAllWindows()
```

```python
cv2.imwrite("final_img.jpg",cphoto4)
```
